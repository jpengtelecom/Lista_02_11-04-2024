```python
import numpy as np
import matplotlib.pyplot as plt
N = 1000000
#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
A = 3
ygx = (((3*y)/A))**(1/3)
#calcular a integral
fx = (((1/ygx - 1)*(1+(((1/ygx - 1))**2))**(-2)))/(ygx**2)
gx = A*(ygx**2)
integral = np.sum(fx/gx) / N
print(f'Integral = {integral:.4f}')

#plotar o grafico
x = np.linspace(0, 30, 500)
curve_y = x*((1 + x**2)**(-2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(5, 0.3,f'Area = {integral:.3f}', fontsize=14)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
