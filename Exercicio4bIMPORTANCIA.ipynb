```python
import numpy as np
import matplotlib.pyplot as plt
N = 1000000
#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
ygx = np.log(y + (np.exp(-2)))
A = 1/(np.exp(2) - np.exp(-2))
#calcular a integral
integral = np.sum((np.exp(ygx + ygx**2)) /(A*np.exp(ygx))) / N
print(f'Integral = {integral:.4f}')
#plotar o grafico
x = np.linspace(-2, 2, N)
curve_y = np.exp((x + x**2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(-1, 200, f'Area = {integral:.4f}', fontsize=14, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
