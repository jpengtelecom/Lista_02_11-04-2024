```python
import numpy as np
import matplotlib.pyplot as plt
N = 100000
#gerar N amostras uniformes
y = np.random.uniform(0, 1, N)
#Gerar as amostras de g(x)
ygx = y**0.4
#calcular a integral
integral = np.sum(((((1-(ygx**2))**1.5)))/((5/2)*ygx**1.5)) / N
print(f'Integral = {integral:.4f}')
#plotar o grafico
x = np.linspace(0, 1, N)
curve_y = (((1-(x**2))**1.5))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='blue')
plt.text(0.5, 0.8, f'Area = {integral:.4f}', fontsize=14, color='red')
plt.xlim(-0.01, 1.1)
plt.ylim(-0.01, 1.1)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
