```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000 #número de amostras
x = np.random.uniform(0, 1, N) #variavel aleatoria

cont_01 = 0
cont_02 = 0
cont_03 = 0
cont_04 = 0

quant = 0

for val in x:

    if val <= 0.25:  # Bola Azul
        cont_01 += 1

    elif val <= 0.5:  # Bola Amarela
        cont_02 += 1

    elif val <= 0.75:  # Bola Vermelha
        cont_03 += 1

    elif val <= 1:  # Bola Roxa
        cont_04 += 1

    cont_00 = cont_01 + cont_02 + cont_03 + cont_04

    if cont_00 == 10:
        if cont_01 == 2 and cont_04 == 2:

            quant += 1

            cont_01 = 0
            cont_02 = 0
            cont_03 = 0
            cont_04 = 0
        else:  # Reiniciar as contagens

            cont_01 = 0
            cont_02 = 0
            cont_03 = 0
            cont_04 = 0

prob = (quant / N) * 100  # Cálculo da Probabilidade
print("A probabilidade : ", prob, "%")


```
